```python
#Determining if Data is Normally Distributed
import scipy.stats as stats
from scipy.stats import shapiro, kstest

# Function to perform Shapiro-Wilk and Kolmogorov-Smirnov tests
def test_normality(data):
    shapiro_test = shapiro(data)
    ks_test = kstest(data, 'norm', args=(np.mean(data), np.std(data)))
    return {
        "Shapiro-Wilk": {"statistic": shapiro_test.statistic, "p-value": shapiro_test.pvalue},
        "Kolmogorov-Smirnov": {"statistic": ks_test.statistic, "p-value": ks_test.pvalue}
    }

# Visualize data with histograms and Q-Q plots
def visualize_distribution(gene, condition, normalized_gene_expression):
    plt.figure(figsize=(12, 5))

    # Histogram with KDE
    plt.subplot(1, 2, 1)
    sns.histplot(normalized_gene_expression, kde=True, bins=10, color='blue')
    plt.title(f"Histogram of {gene} - {condition}")
    plt.xlabel("Normalized Gene Expression")
    plt.ylabel("Frequency")

    # Q-Q Plot
    plt.subplot(1, 2, 2)
    stats.probplot(normalized_gene_expression, dist="norm", plot=plt)
    plt.title(f"Q-Q Plot of {gene} - {condition}")

    plt.tight_layout()
    plt.show()

# Analyze each gene and condition
for gene, conditions in data.items():
    for condition, normalized_gene_expression in conditions.items():
        print(f"Testing normality for {gene} - {condition}")

        # Normality tests
        results = test_normality(normalized_gene_expression)
        print(f"Shapiro-Wilk Test: W={results['Shapiro-Wilk']['statistic']:.4f}, p-value={results['Shapiro-Wilk']['p-value']:.4f}")
        print(f"Kolmogorov-Smirnov Test: D={results['Kolmogorov-Smirnov']['statistic']:.4f}, p-value={results['Kolmogorov-Smirnov']['p-value']:.4f}")
        print("-" * 50)

        # Visualization
        visualize_distribution(gene, condition, normalized_gene_expression)
```

```python
#Power Analysis

import numpy as np
from statsmodels.stats.power import TTestIndPower

# Function to calculate Cohen's d (effect size)
def cohen_d(group1, group2):
    # Check that the groups are numpy arrays
    group1 = np.array(group1)
    group2 = np.array(group2)

    # Check if both groups have the same length
    if len(group1) != len(group2):
        raise ValueError("Both groups must have the same length")

    # Calculate means and standard deviations
    mean1, mean2 = np.mean(group1), np.mean(group2)
    std1, std2 = np.std(group1, ddof=1), np.std(group2, ddof=1)

    # Calculate pooled standard deviation
    pooled_std = np.sqrt(((len(group1) - 1) * std1**2 + (len(group2) - 1) * std2**2) / (len(group1) + len(group2) - 2))

    # Calculate Cohen's d
    d = (mean1 - mean2) / pooled_std
    return d

# Function to perform power analysis given Cohen's d
def power_analysis(effect_size, alpha=0.05, power=0.80):
    analysis = TTestIndPower()
    sample_size = analysis.solve_power(effect_size=effect_size, alpha=alpha, power=power, alternative='two-sided')
    return sample_size

# Extract the groups from the data dictionary
group1 = data["ARC"]["Control Normoxia"]
group2 = data["ARC"]["Control Hypoxia"]

# Calculate Cohen's d (effect size)
effect_size = cohen_d(group1, group2)

# Perform power analysis to find the sample size needed for 80% power
sample_size = power_analysis(effect_size)

# Output the results
print(f"Cohen's d (effect size) for Control Normoxia vs. Control Hypoxia (ARC): {effect_size:.4f}")
print(f"Sample size needed per group for 80% power and alpha = 0.05: {np.ceil(sample_size)}")
```
